```python
def svd_dot_u (array1, array2):
  svd_dot_u=[(np.dot(array1[i],array2[i].T)) for i in range (array1.shape[0])]
  return svd_dot_u
```

```python
def svd_dot_v (array1, array2):
  svd_dot_v=[(np.dot(array1[i],array2[i].T)) for i in range (array1.shape[0])]
  return svd_dot_v
```

```python
df_final_train['svd_dot_u']=svd_dot_u(svd_u_s_train,svd_u_d_train)
df_final_train['svd_dot_v']=svd_dot_v(svd_v_s_train,svd_v_d_train)
df_final_test['svd_dot_u']=svd_dot_u(svd_u_s_test,svd_u_d_test)
df_final_test['svd_dot_v']=svd_dot_v(svd_v_s_test,svd_v_d_test)
```

```python
x_train=df_final_train.copy()
x_test=df_final_test.copy()
```

```python
x_train.drop(['source_node', 'destination_node','indicator_link'],axis=1,inplace=True)
x_test.drop(['source_node', 'destination_node','indicator_link'],axis=1,inplace=True)
```

```python
y_train=df_final_train['indicator_link']
y_test=df_final_test['indicator_link']
```

```python
estimator_list=[30,60,100,200,300,400]
subsample_list=[0.4,0.7]
depth_list=[3,7,12]
train_value=[]
test_value=[]
#we will not use column sampling as the number of features are very limited.
for estimator in tqdm(estimator_list):
  for sample in tqdm(subsample_list):
    for depth in tqdm(depth_list):
      clf =  XGBClassifier(n_estimators=estimator,max_depth=depth,subsample=sample,n_jobs=-1)
      clf.fit(x_train,y_train)
      train_labels=clf.predict(x_train)
      test_labels=clf.predict(x_test)
      train_score=f1_score(y_train,train_labels)
      test_score=f1_score(y_test,test_labels)
      train_value.append(train_score)
      test_value.append(test_score)
print(train_value)
print(test_value)

```

```python
mylist=[]
for i in range (len(train_value)):
    diff=train_value[i]-test_value[i]
    mylist.append(diff)
# mylist
```

```python
best_estimator=30
best_subsample=0.7
best_depth=7
best_model=XGBClassifier(n_estimators=best_estimator,max_depth=best_depth,subsample=best_subsample,random_state=0,n_jobs=-1)
best_model.fit(x_train,y_train)
train_labels_best=best_model.predict(x_train)
test_labels_best=best_model.predict(x_test)
f1_train=f1_score(y_train,train_labels_best)
f1_test=f1_score(y_test,test_labels_best)

print("The train F1 score is",f1_train)
print("The test F1 score is",f1_test)
```

```python
from sklearn.metrics import confusion_matrix
def plot_confusion_matrix(test_y, predict_y):
    C = confusion_matrix(test_y, predict_y)
    
    A =(((C.T)/(C.sum(axis=1))).T)
    
    B =(C/C.sum(axis=0))
    plt.figure(figsize=(20,4))
    
    labels = [0,1]
    # representing A in heatmap format
    cmap=sns.light_palette("blue")
    plt.subplot(1, 3, 1)
    sns.heatmap(C, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Confusion matrix")
    
    plt.subplot(1, 3, 2)
    sns.heatmap(B, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Precision matrix")
    
    plt.subplot(1, 3, 3)
    # representing B in heatmap format
    sns.heatmap(A, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Recall matrix")
    
    plt.show()
```

```python
print('Train confusion_matrix')
plot_confusion_matrix(y_train,train_labels_best)
print('Test confusion_matrix')
plot_confusion_matrix(y_test,test_labels_best)
```

```python
from sklearn.metrics import roc_curve, auc
fpr,tpr,ths = roc_curve(y_test,test_labels_best)
auc_sc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='navy',label='ROC curve (area = %0.2f)' % auc_sc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic with test data')
plt.legend()
plt.show()
```
